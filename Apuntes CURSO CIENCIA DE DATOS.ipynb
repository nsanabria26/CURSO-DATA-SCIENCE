**Variables y tipos de dato**

Una variable es un lugar en la memoria que va a ser usado para guardar un valor concreto


```python
nombre = "Nahuel"
print(nombre)
nombre = "Facundo"
print(nombre)
```

Todo lo que printee entre comillas se va a imprimir de forma literal

Si hay un solo print, va a imprimir el valor que fue escrito previamente al comando

Pyton siempre se queda con el ultimo valor que fue asignado a la variable #OVERWRITING

```python
print("Hola a todos, dijo",nombre) #CONCATENAR #Es mezclar texto con variables
```

**CASTEO DE VARIABLES**

Convertir un tipo de datos a otro distinto

```python
temperatura = int(temperatura) # redefino el tipo y le asigno a la variable temperatura, un numero entero
```

```python
print(type(temperatura))    #int es una palabra reservada para convertir el tipo de dato de una variable a un numero entero
```

```python
if temperatura > 20:
  print("Está para pileta")
else:                               #En caso de que no se cumpla la primer condicion, se ejecutara el siguiente bloque
  print("No está para pileta")
```
